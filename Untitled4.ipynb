```python
# ==============================
# TASK 2: EDA (EXPLORATORY DATA ANALYSIS)
# ==============================

# Step 1: Upload cleaned dataset
from google.colab import files
uploaded = files.upload()

# Step 2: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 3: Load dataset
file_name = next(iter(uploaded))
df = pd.read_csv(file_name)

print("✅ Dataset Loaded:", file_name)

# Step 4: Basic Info
print("\n🔹 Shape:", df.shape)
print("\n🔹 Columns:", df.columns)
print("\n🔹 Info:")
df.info()

# Step 5: Statistical Summary
print("\n🔹 Statistical Summary:")
print(df.describe())

# Step 6: Correlation Heatmap
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()

# Step 7: Distribution Plots (Numeric Columns)
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()

# Step 8: Count Plots (Categorical Columns)
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    plt.figure()
    sns.countplot(x=df[col])
    plt.title(f"Count Plot of {col}")
    plt.xticks(rotation=45)
    plt.show()

print("\n🎉 TASK 2 (EDA) COMPLETED!")
```
